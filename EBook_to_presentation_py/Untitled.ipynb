```python
import pptx

# create a new PowerPoint presentation
prs = pptx.Presentation()

# add a new slide with a title and a bullet point list
slide = prs.slides.add_slide(prs.slide_layouts[1])
title = slide.shapes.title
title.text = 'My Presentation'
subtitle = slide.placeholders[1]
subtitle.text = 'Subtitle'

bullet_points = ['Bullet 1', 'Bullet 2', 'Bullet 3']
left = top = pptx.util.Inches(2)
width = pptx.util.Inches(6)
height = pptx.util.Inches(1.5)
txBox = slide.shapes.add_textbox(left, top, width, height)
tf = txBox.text_frame

tf.text = 'Bullet Points:'
for point in bullet_points:
    p = tf.add_paragraph()
    p.text = point
    p.level = 1

# save the presentation to a file
prs.save('my_presentation3.pptx')
```

```python
import win32com.client

# Open a PowerPoint presentation
powerpoint = win32com.client.Dispatch("PowerPoint.Application")
presentation = powerpoint.Presentations.Open("example1.pptx")

# Add an image to the first slide
slide = presentation.Slides(1)
picture_path = "dht11.png"
picture = slide.Shapes.AddPicture(picture_path, LinkToFile=False, SaveWithDocument=True, Left=0, Top=0)

# Add a video to the second slide
slide = presentation.Slides(2)
video_path = "1.mp4"
video = slide.Shapes.AddMediaObject2(video_path, False, True, Left=0, Top=0, Width=640, Height=480)

# Close the presentation
presentation.Save()
presentation.Close()

# Quit PowerPoint
powerpoint.Quit()
```
